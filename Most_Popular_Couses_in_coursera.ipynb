## Step 4. Preprocessing

```
# level column 확인
df['level'].unique()
```

```
# level column을 숫자형태로 변형합니다.
def levels(x):
    if x == 'Beginner':
        return 1
    elif x == 'Mixed':
        return 2
    elif x == 'Intermediate':
        return 3
    elif x == 'Advanced':
        return 4
```

```
df['level'] = df['level'].apply(levels)
df['level'].unique()
```

```
# enrolled column 확인
df['enrolled']
```

```
# 단위(k)를 숫자로 변형합니다.
def counts(x):
    rx = x.replace('k','000')
    if '.' in rx:
        rx = rx.replace('.','')
        rx = rx[:-1]
        return int(rx)
    return int(rx)
```

```
# counts()함수가 잘 동작하는지 테스트합니다.
df['enrolled'].apply(counts)
```

```
# 데이터 프레임에 적용합니다.
df['enrolled'] = df['enrolled'].apply(counts)
df
```
